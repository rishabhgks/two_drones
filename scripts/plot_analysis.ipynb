```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ber = pd.read_csv("/home/rishabh/ber.csv")
ber.head()
```

```python
uas5pos = pd.read_csv("/home/rishabh/uas5_pos.csv")
uas5pos.head()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(20,10))
ax1.plot(ber['%time'], ber['field.data'])
ax1.set_xlabel("Time")
ax1.set_ylabel("BER")
ax2.plot(uas9pos['%time'], uas9pos['field.pose.pose.position.x'])
ax2.set_xlabel("Time")
ax2.set_ylabel("UAS 1 X-Axis Position")
ax3.plot(uas5pos['%time'], uas5pos['field.pose.pose.position.x'])
ax3.set_xlabel("Time")
ax3.set_ylabel("UAS 2 X-Axis Position")
fig.savefig("xaxis_plot.png")
```
